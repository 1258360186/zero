```python
import pandas as pd
import numpy as np
scenery_file_path = open('.csv')
scenery_data = pd.read_csv(scenery_file_path)
scenery_data
```

```python
area = float("{:.1f}".format(
scenery_data['()'].mean()))
tourist = float("{:.1}".format(
scenery_data['()'].mean()))
values = {"()":area,"()":tourist}
scenery_data = scenery_data.fillna(value=values)
scenery_data.head
```

```python
data = scenery_data.groupby('')
hebei_scenery = dict([x for x in data])['']
hebei_scenery
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']  
plt.rcParams['axes.unicode_minus']=False    
area = hebei_scenery['()'].values
tourist = hebei_scenery['()'].values
plt.figure(figsize=(12, 6))
x_num = range(0, len(area))
x_dis = [i + 0.3 for i in x_num]
plt.bar(x_num, area, color='g', width=.3, label='')
plt.bar(x_dis, tourist, color='r', width=.3, label='')
plt.ylabel('/')
plt.title('')
plt.legend(loc='upper right')
plt.xticks(range(0, 10),['', '', '-', 
'','','','',
'','',''])
plt.show()
```

```python
import matplotlib.pyplot as plt
every_scenery = hebei_scenery['()'].values
all_scenery = hebei_scenery['()'].sum()
percentage = (every_scenery/all_scenery)*100
np.set_printoptions(precision=2)
labels  = ['', '', '-', 
            '','','','',
            '','','']
plt.axes(aspect=1)
plt.pie(x= percentage, labels=labels, autopct='%3.2f %%',shadow=True, labeldistance=1.2, startangle = 90,pctdistance = 0.7)
plt.legend(loc='upper left')
plt.show()
```
